```python
#What was the daily return average of a stock? 
#The Daily Return Average of a stock
data_amazon['Return'] = (data_amazon['Close']/data_amazon['Close'].shift(1))-1


```

```python
data_amazon
data_amazon['Return'].isnull()
```

```python
data_amazon['Return'].mean()
```

```python
data_amazon
```

```python
data_microsoft['Return'] = (data_microsoft['Close']/data_microsoft['Close'].shift(1))-1
data_google['Return'] = (data_google['Close']/data_google['Close'].shift(1))-1
data_apple['Return'] = (data_apple['Close']/data_apple['Close'].shift(1))-1
```

```python
corpus=pd.concat([data_amazon['Return'],data_google['Return'],data_apple['Return'],data_microsoft['Return']],axis=1)
corpus.columns={"Amazon Return","Google Return","Apple Return","Microsoft Return"}
```

```python
corpus.dropna(inplace=True)
```

```python
corpus
```

```python
corpus.corr()
```

```python
import seaborn as sns
cor=corpus.corr()
plt.figure(figsize=(15,7))
sns.heatmap(data=cor,annot=True)
```

```python
import numpy as np
data_amazon["Trends"] =  np.where(data_amazon["Return"] >= 0 , "Positive", "Negative")
```

```python
data_amazon
```

```python
data_amazon.Trends.value_counts().plot(kind='pie',labels=['positive','negative'])
plt.axis('equal')
plt.title('Trend Frequency')
```
